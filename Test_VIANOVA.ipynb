```python
import csv
import sqlite3
#Connecting to the database
conn = sqlite3.connect('cities.db')
cursor = conn.cursor()

# Creating a city table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS cities (
        country_code TEXT,
        country_name TEXT,
        city_name TEXT,
        population INTEGER
    )
''')

with open('new_table.csv', 'r', encoding='utf-8') as file:
    reader = csv.reader(file)
    next(reader)  # Skip the title line
    for row in reader:
        country_code = row[1]
        country_name = row[2]
        city_name = row[0]
        population = int(row[3])

        cursor.execute('''
            INSERT INTO cities (country_code, country_name, city_name, population)
            VALUES (?, ?, ?, ?)
        ''', (country_code, country_name, city_name, population))

# Submit and close the connection
conn.commit()
conn.close()
```

```python
import csv
import sqlite3

# Connecting to the database
conn = sqlite3.connect('cities.db')
cursor = conn.cursor()

# Execute SQL queries and store the results in a DataFrame
query = "SELECT country_code, country_name FROM (SELECT country_code, country_name, MAX(population) AS max_population FROM cities GROUP BY country_code, country_name) WHERE max_population < 10000000 ORDER BY country_name ASC"
df = pd.read_sql_query(query, conn)

# Save the DataFrame as a CSV file
df.to_csv('result.csv', index=False,encoding='utf-8')

conn.close()
```
